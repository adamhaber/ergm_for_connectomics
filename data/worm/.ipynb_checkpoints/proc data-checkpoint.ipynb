```python
df = pd.read_excel("ei classification.xlsx", sheet_name="Hermaphrodite, sorted by NT",)
```

```python
df = df.iloc[5:,:][["Unnamed: 3","Unnamed: 5"]]
```

```python
df.columns = ['Neuron', 'transmitter']
```

```python
df.transmitter.fillna('NA').replace("Unknown","UNK")

df['transmitter']=df['transmitter'].apply(
    lambda x: x if x in ["Acetylcholine (ACh)","Glutamate (Glu)","GABA"] else "E")
```

```python
df.to_csv("transmitters.csv")
```
